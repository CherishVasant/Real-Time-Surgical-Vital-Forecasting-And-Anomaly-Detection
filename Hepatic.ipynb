# Loading library and data

To use the VitalDB open dataset, the pandas library is required.

At first, we need to load 3 endpoints of the VitalDB open dataset. 

```python
import numpy as np
import pandas as pd

df_cases = pd.read_csv("https://api.vitaldb.net/cases")  # clinical information
df_trks = pd.read_csv("https://api.vitaldb.net/trks")  # track list
df_labs = pd.read_csv('https://api.vitaldb.net/labs')  # laboratory results
```

# Hepatic

```python
hepatic_cases = df_cases[df_cases['optype'] == 'Hepatic']
hepatic_cases
```

```python
hepatic_caseids = hepatic_cases['caseid'].tolist()
hepatic_caseids
```

```python
# Load track data for hepatic cases
print("📈 Loading track data for hepatic cases...")

# Get tracks available for hepatic cases
hepatic_tracks = df_trks[df_trks['caseid'].isin(hepatic_caseids)]
print(f"Found {len(hepatic_tracks)} track records for {len(hepatic_caseids)} hepatic cases")
print(f"Unique track types available: {hepatic_tracks['tname'].nunique()}")

# Show track categories for hepatic cases
print("\n🔍 Track categories for hepatic cases:")
track_categories = {}
for track in hepatic_tracks['tname'].unique():
    device = track.split('/')[0] if '/' in track else 'Unknown'
    if device not in track_categories:
        track_categories[device] = []
    track_categories[device].append(track)

for device, tracks in sorted(track_categories.items()):
    print(f"\n{device} ({len(tracks)} tracks):")
    for track in sorted(tracks):
        print(f"  - {track}")

# Show most common tracks for hepatic cases
print(f"\n📊 Most common tracks for hepatic cases:")
track_counts = hepatic_tracks['tname'].value_counts().head(15)
for track, count in track_counts.items():
    percentage = (count / len(hepatic_caseids)) * 100
    print(f"  - {track}: {count} cases ({percentage:.1f}%)")
```

```python
# Load lab data for hepatic cases
print("🧪 Loading lab data for hepatic cases...")

# Get lab results for hepatic cases
hepatic_labs = df_labs[df_labs['caseid'].isin(hepatic_caseids)]
print(f"Found {len(hepatic_labs)} lab records for hepatic cases")
print(f"Unique lab types available: {hepatic_labs['name'].nunique()}")

# Show lab types and their frequency for hepatic cases
print(f"\n🔬 Lab types for hepatic cases:")
lab_counts = hepatic_labs['name'].value_counts()
for lab, count in lab_counts.items():
    percentage = (count / len(hepatic_caseids)) * 100
    print(f"  - {lab}: {count} records ({percentage:.1f}% of cases)")

# Show key liver function tests
print(f"\n🫀 Key liver function tests for hepatic cases:")
liver_function_tests = ['ast', 'alt', 'tbil', 'alb', 'tprot', 'ptsec', 'ptinr', 'aptt', 'fib']
for test in liver_function_tests:
    if test in hepatic_labs['name'].values:
        count = (hepatic_labs['name'] == test).sum()
        percentage = (count / len(hepatic_caseids)) * 100
        print(f"  - {test}: {count} records ({percentage:.1f}% of cases)")
    else:
        print(f"  - {test}: Not available")

# Show basic metabolic panel
print(f"\n🧪 Basic metabolic panel for hepatic cases:")
metabolic_tests = ['na', 'k', 'cl', 'cr', 'bun', 'gfr', 'gluc']
for test in metabolic_tests:
    if test in hepatic_labs['name'].values:
        count = (hepatic_labs['name'] == test).sum()
        percentage = (count / len(hepatic_caseids)) * 100
        print(f"  - {test}: {count} records ({percentage:.1f}% of cases)")
    else:
        print(f"  - {test}: Not available")
```

```python
# Load waveform data for hepatic cases using vitaldb
import vitaldb

print("🔄 Loading waveform data for hepatic cases...")

# Get available track names for hepatic cases
available_tracks = hepatic_tracks['tname'].unique().tolist()
print(f"Available tracks: {len(available_tracks)} types")

# Select key tracks for hepatic surgery monitoring
key_tracks = [
    'Solar8000/HR',           # Heart rate
    'Solar8000/ART_SBP',      # Systolic blood pressure  
    'Solar8000/ART_DBP',      # Diastolic blood pressure
    'Solar8000/ART_MBP',      # Mean blood pressure
    'Solar8000/PLETH_SPO2',   # Oxygen saturation
    'SNUADC/ART',             # Arterial pressure waveform
    'SNUADC/ECG_II',          # ECG Lead II
    'Primus/CO2',             # CO2 monitoring
    'Primus/ETCO2',           # End-tidal CO2
    'BIS/BIS'                 # Bispectral index
]

# Filter to only tracks that are available for hepatic cases
available_key_tracks = [track for track in key_tracks if track in available_tracks]
print(f"Key tracks available: {len(available_key_tracks)}")

if len(available_key_tracks) > 0:
    print("Available key tracks:")
    for track in available_key_tracks:
        print(f"  - {track}")
        
    # Load waveform data for first hepatic case as example
    if len(hepatic_caseids) > 0:
        example_case_id = hepatic_caseids[0]
        print(f"\n📊 Loading waveform data for hepatic case {example_case_id}...")
        
        try:
            # Load waveform data
            waveform_data = vitaldb.load_case(
                caseid=example_case_id,
                track_names=available_key_tracks,
                interval=1/100  # 100Hz sampling rate
            )
            
            print(f"✅ Successfully loaded waveform data!")
            print(f"  - Data shape: {waveform_data.shape}")
            print(f"  - Time points: {waveform_data.shape[0]}")
            print(f"  - Tracks: {waveform_data.shape[1]}")
            
            # Show data availability for each track
            print("\n📈 Data availability for each track:")
            for i, track_name in enumerate(available_key_tracks):
                track_data = waveform_data[:, i]
                non_nan_count = np.sum(~np.isnan(track_data))
                total_count = len(track_data)
                availability = (non_nan_count / total_count) * 100
                print(f"  - {track_name}: {non_nan_count}/{total_count} ({availability:.1f}%)")
                
        except Exception as e:
            print(f"❌ Error loading waveform data: {e}")
else:
    print("❌ No key tracks available for hepatic cases")
```
